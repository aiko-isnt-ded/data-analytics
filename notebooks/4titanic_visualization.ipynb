# Titanic Data Visualization

```python
import pandas as pd
import matplotlib.pyplot as plt
path = "C:/Users/Viviana/apps/data-analytics/data/raw/test.csv"
```

```python
df = pd.read_csv(path)
df.head()
```

### General Statistical Measures of the Data

```python
df.describe()
```

```python
df["Pclass"].value_counts()
```

There were 418 passengers aboard: 107 of them were upper class, 93 middle class, and 218 of them were lower class.

The average passenger aged 30, and their ages ranged from 0 to 70 years-old.

The average ticket fare was £35 pounds, and the maximum amount paid was £512.

### General Data Analysis

```python
df["Pclass"].value_counts().plot(kind="bar", title="Passengers by Class", xlabel="Class", colormap="Accent")
```

The passengers travelling in the Titanic were mostly from the lower class, twice the amount as the other groups.

```python
df["Sex"].value_counts().plot(kind="bar", title="Passengers by Sex", xlabel="Sex", colormap="tab20b")
```

The passengers aboard the Titanic were mostly men, doubling the amount of women in the ship.

```python
df["Age"].plot(kind="hist", title="Passengers by Age", xlabel="Age", colormap="plasma")
```

Most of the passengers travelling in the Titanic aged 20 to 30 years-old. There were also more children than elderly. 

```python
df["SibSp"].plot(kind="hist", title="Siblings to Spouses Ratio", colormap="viridis")
```

There were more spouses than siblings aboard the Titanic ship, as shown by the data concentration in the range 0 to 1. 

If the opposite were true (more spouses than siblings), the numerator would overwhelm the denominator, resulting in bigger values.

```python
df["Parch"].plot(kind="hist", title="Parents to Children Ratio", colormap="magma")
```

The value of 0 in this graphs stands for those children who were not accompanied by their parents. The large concentration of data in this range shows that most of the children aboard the ship travelled with nannys or other people.

Thus, there were more children than parents aboard the ship.

```python
df["Fare"].plot(kind="hist", title="Ticket Fare", colormap="cividis")
```

Most of the passengers aboard the Titanic purchased their tickets below £100 pounds. This represents more than 350 people. 

### Data Analysis by Class

```python
# Defining the classes
upper = df[df["Pclass"] == 1]
middle = df[df["Pclass"] == 2]
lower = df[df["Pclass"] == 3]
```

```python
upper["Fare"].plot(kind="hist", title="Ticket fare paid by the upper class", colormap="Pastel1", xlabel="Ticket Fare",bins=20)
```

As shown by the graph above, most of the passengers travelling in upper class paid around £25 to £75 for their tickets in the Titanic.
There’s also a possible outlier, who paid £500 for their ticket. 

```python
middle["Fare"].plot(kind="hist", title="Ticket fare paid by the middle class", cmap="Pastel2", xlabel="Ticket Fare",bins=20)
```

In comparison to the tickets bought by the upper class, the middle class ticket fares do not exceed £100 pounds. The data is concentrated around the £10 to £20 range too. This suggests that the middle class paid significantly lower amounts for their tickets. 

```python
lower["Fare"].plot(kind="hist", title="Ticket fare paid by the lower class", colormap="Paired", xlabel="Ticket Fare",bins=20)
```

Finally, the majority of the lower class bought their tickets for less than £10 pounds. They also are the class with most passengers (218) aboard the Titanic. 

```python
# Create a grid for the 3 plots
fig, axes = plt.subplots(nrows=1, ncols=3)

upper["Age"].plot(ax=axes[0], title="Upper Class", kind="hist", bins=20, xlabel="Age", colormap="tab10", figsize=(12,3), xlim=(0,80))
lower["Age"].plot(ax=axes[1], title="Middle Class", kind="hist", bins=20, xlabel="Age", ylabel="", colormap="Dark2", xlim=(0,80))
middle["Age"].plot(ax=axes[2], title="Lower Class", kind="hist", bins=20, xlabel="Age", ylabel="", colormap="Set3", xlim=(0,80))
```

By analizing the ages distributed along the upper, middle and lower classes:

The ages of the upper class range from 6 to 76, and show a greater concentration in the middle range (30 to 50 years-old). This is the only class in which ages exceed 65.

The ages of the middle class range from 0 to 63, and show a greater concentration in the range from (roughly) 18 to 25. This class has the most variety in the ages of children, and holds the biggest amount of newly-born babies. 

The ages of the lower class range from 0 to 60, and show a greater concentration in the range from 20 to 30. This suggests that most of the passengers were in working age. 

### Data Analysis by Sex

```python
fig, axes = plt.subplots(nrows=1, ncols=3)

upper["Sex"].value_counts().plot(ax=axes[0], title="Upper Class", kind="bar", colormap="tab10", figsize=(12,3))
middle["Sex"].value_counts().plot(ax=axes[1], title="Middle Class", kind="bar", colormap="Dark2", figsize=(12,3))
lower["Sex"].value_counts().plot(ax=axes[2], title="Lower Class", kind="bar", colormap="Set2", figsize=(12,3))
```

As shown by the bar graphs, the upper class had the most female passengers among the other groups, and the shortest gap between sexes. 

Meanwhile, both the middle and lower classes had a bigger gap between their female and male passengers.
